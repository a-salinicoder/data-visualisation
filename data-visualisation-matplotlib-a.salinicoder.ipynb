# Data Visualization in Python

## Importance of Data Visualization
- **Understanding Data**: Helps in understanding the patterns, trends, and insights in large datasets.
- **Storytelling**: Aids in telling a story by visualizing the data in a form that is easy to understand and interpret.
- **Decision Making**: Assists stakeholders and decision-makers to make informed decisions based on visual insights.
- **Exploratory Data Analysis (EDA)**: Plays a crucial role in exploratory data analysis, where visualizations help in identifying anomalies, outliers, and relationships between variables.

## Libraries for Data Visualization in Python
- **Matplotlib**: A flexible and widely used library for creating static, interactive, and animated visualizations in Python.
- **Seaborn**: Built on top of Matplotlib, Seaborn provides a high-level interface for drawing attractive and informative statistical graphics.
- **Plotly**: A library for creating interactive plots. It supports a wide range of chart types and can be used in Jupyter Notebooks, web applications, and more.
- **Bokeh**: Aims to provide elegant, concise construction of versatile graphics and to extend this capability with high-performance interactivity over large or streaming datasets.
- **Pandas Plotting**: Pandas, the popular data manipulation library, also includes functionalities for simple plotting based on Matplotlib.
- **Altair**: A declarative statistical visualization library that is user-friendly and generates visually appealing charts.

## Basic Steps in Data Visualization
1. **Data Cleaning and Preparation**: Before visualization, it is important to clean and prepare the data to avoid any discrepancies in the visualizations.
2. **Selection of Appropriate Chart Type**: Depending on the data and the insights you want to derive, select the most appropriate chart type, e.g., bar chart, line chart, scatter plot, etc.
3. **Customization**: Customize the chart by adding labels, titles, colors, etc., to make it more informative.
4. **Interactivity (Optional)**: Add interactivity to the charts (using libraries like Plotly or Bokeh) to provide a more immersive experience for the viewer.
5. **Interpretation**: Interpret the visualization to derive insights and conclusions.

## Common Types of Visualizations
- **Bar Chart**: Used to show comparisons among discrete categories.
- **Line Chart**: Ideal for showing trends over a continuous interval or time span.
- **Scatter Plot**: Used to show relationships between two numerical variables.
- **Histogram**: Represents the distribution of a continuous variable.
- **Pie Chart**: Useful for showing proportions of a whole.
- **Heatmap**: Shows the magnitude of a phenomenon as color in two dimensions.
- **Box Plot**: Depicts groups of numerical data through their quartiles.

## Best Practices in Data Visualization
- **Simplicity**: Keep the visualizations simple and easy to understand.
- **Clarity**: Make sure the visualization conveys the message clearly without any ambiguity.
- **Labeling**: Always label the axes, legends, and other components to provide context.
- **Colors**: Use colors judiciously to highlight important points or to represent different categories.
- **Interactivity**: When necessary, add interactivity to enhance the user experience.
- **Storytelling**: Focus on storytelling to convey the message more effectively.

# Matplotlib: A Comprehensive Guide

## What is Matplotlib?

Matplotlib is a plotting library for Python and its numerical mathematics extension, NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK.

## Why Use Matplotlib?

- **Versatility**: Offers a wide range of plot types, from histograms and scatter plots to non-Cartesian coordinates graphs.
- **Customization**: Allows for extensive customization to tailor plots to specific needs.
- **Ease of Use**: Simple syntax for creating plots quickly, as well as advanced features for expert users.
- **Integration**: Integrates seamlessly with NumPy for numerical analysis.

## Key Concepts

- **Figure**: The whole figure which includes everything on the page, such as axes, graphics, text, and labels.
- **Axes**: The area on which data is plotted, it includes an x-axis and a y-axis.
- **Axis**: These are the number-line-like objects, they take care of generating the graph limits.
- **Artist**: Everything on the figure is an artist, including text, line2D, and collection objects.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv('/Users/salinianbalagan/Desktop/Data Visualisation - Salini/titanic-dataset.csv')
````

````python
# Identifying missing values
missing_values = data.isnull().sum()
print("Missing Values in Each Column:\n", missing_values)
````

````python
# Exploring data types
print("Data Types of Each Column:")
data.info()
````

````python
#Bar Graph

# Bar graph of survivors
survivors = data['Survived'].value_counts()
survivors.plot(kind='bar', color=['skyblue', 'lightcoral'])

plt.title('Survivors on the Titanic')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Did not survive', 'Survived'], rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
# Pie chart of gender

gender = data['Sex'].value_counts()
gender.plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral', 'skyblue'], textprops={'fontsize': 12})
plt.title('Gender Distribution on the Titanic')
plt.ylabel('')
plt.show()
````

````python
# Histogram of ages

data['Age'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
# Scatter plot of age vs fare

plt.scatter(data['Age'], data['Fare'], alpha=0.5, c='skyblue', s=10)
plt.title('Age vs Fare on the Titanic')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.show()
````

````python
# Box plot of ages

boxplot = data['Age'].plot(kind='box', color='skyblue')
plt.title('Age Distribution on the Titanic')
plt.ylabel('Age')
for line in boxplot.lines:
    line.set_linewidth(2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Box plot of fares

boxplot = data['Fare'].plot(kind='box', color='lightcoral')
plt.title('Fare Distribution on the Titanic')
plt.ylabel('Fare')
for line in boxplot.lines:
    line.set_linewidth(2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python
# Violin plot of ages
sns.violinplot(data['Age'], color='skyblue')
plt.title('Age Distribution on the Titanic')
plt.xlabel('Age')
plt.show()

# Violin plot of fares
sns.violinplot(data['Fare'], color='lightcoral')
plt.title('Fare Distribution on the Titanic')
plt.xlabel('Fare')
plt.show()
````

````python
# Heatmap of correlation matrix

correlation = data.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Titanic Dataset')
plt.show()
````

````python
#Stack Plot


# Group by Pclass and Survived and count the number of occurrences
grouped = data.groupby(['Pclass', 'Survived']).size().unstack()

# Normalize the data to get the proportion of survivors in each class
grouped = grouped.div(grouped.sum(axis=1), axis=0)

# Plot the stacked bar plot
grouped.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'])
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Proportion')
plt.legend(['Did not Survive', 'Survived'])
plt.show()
````

````python
# Create a figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

# Plot Age distribution on the first subplot
ax[0, 0].hist(data['Age'].dropna(), color='skyblue', edgecolor='black')
ax[0, 0].set_title('Age Distribution')
ax[0, 0].set_xlabel('Age')
ax[0, 0].set_ylabel('Frequency')

# Plot Fare distribution on the second subplot
ax[0, 1].hist(data['Fare'].dropna(), color='lightcoral', edgecolor='black')
ax[0, 1].set_title('Fare Distribution')
ax[0, 1].set_xlabel('Fare')
ax[0, 1].set_ylabel('Frequency')

# Plot Survival rate by class on the third subplot
grouped = data.groupby(['Pclass', 'Survived']).size().unstack()
grouped = grouped.div(grouped.sum(axis=1), axis=0)
grouped.plot(kind='bar', stacked=True, color=['lightcoral', 'skyblue'], ax=ax[1, 0])
ax[1, 0].set_title('Survival Rate by Class')
ax[1, 0].set_xlabel('Passenger Class')
ax[1, 0].set_ylabel('Proportion')

# Plot Embarked distribution on the fourth subplot
data['Embarked'].value_counts().plot(kind='bar', ax=ax[1, 1], color='lightgreen', edgecolor='black')
ax[1, 1].set_title('Embarked Distribution')
ax[1, 1].set_xlabel('Embarked')
ax[1, 1].set_ylabel('Frequency')

# Adjust the spacing between plots
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Count the number of survivors in each passenger class
survivors = data[data['Survived'] == 1]['Pclass'].value_counts().sort_index()

# Create a bar plot
ax = survivors.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Number of Survivors in Each Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Survivors')

# Annotate each bar with the count of survivors
for i, value in enumerate(survivors):
    plt.text(i, value, str(value), ha='center', va='bottom')

# Show the plot
plt.show()
````

### 3. Differences between Seaborn and Matplotlib:
